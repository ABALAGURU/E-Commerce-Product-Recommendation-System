```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('flipkart_com-ecommerce_sample.csv')
```

```python
null_counts = data.isnull().sum()
columns_with_nulls = null_counts[null_counts > 0].index

for column in columns_with_nulls:
    if data[column].dtype == 'object': 
        data[column] = data[column].fillna(data[column].mode()[0])
    
    elif data[column].dtype in ['int64', 'float64']: 
        data[column] = data[column].fillna(data[column].mean()) 
```

```python
data['product_category_tree']=data['product_category_tree'].map(lambda x:x.strip('[]'))
data['product_category_tree']=data['product_category_tree'].map(lambda x:x.strip('"'))
data['product_category_tree']=data['product_category_tree'].map(lambda x:x.split('>>'))
```

```python
unwanted_columns = ['uniq_id', 'crawl_timestamp', 'product_url', 'image', 'is_FK_Advantage_product', 'product_specifications'] 
data = data.drop(unwanted_columns, axis=1)
```

```python
import string
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer 
lem = WordNetLemmatizer()
stop_words = set(stopwords.words('english')) 
exclude = set(string.punctuation)
```

```python
datacopy=data.copy()
datacopy.drop_duplicates(subset ="product_name", 
                     keep = "first", inplace = True)
datacopy.shape
```

```python
import nltk
nltk.download('wordnet', download_dir='/path/to/your/nltk_data')
```

```python
from nltk.corpus import wordnet
```

```python
def filter_keywords(doc):
    doc=doc.lower()
    stop_free = " ".join([i for i in doc.split() if i not in stop_words])
    punc_free = "".join(ch for ch in stop_free if ch not in exclude)
    word_tokens = word_tokenize(punc_free)
    filtered_sentence = [(lem.lemmatize(w, "v")) for w in word_tokens]
    return filtered_sentence
```

```python
datacopy['product'] = datacopy['product_name'].apply(filter_keywords)
datacopy['description'] = datacopy['description'].astype("str").apply(filter_keywords)
datacopy['brand'] = datacopy['brand'].astype("str").apply(filter_keywords)
```

```python
datacopy["all_meta"]=datacopy['product']+datacopy['brand']+ data['product_category_tree']+datacopy['description']
datacopy["all_meta"] = datacopy["all_meta"].apply(lambda x: ' '.join(x))
```
